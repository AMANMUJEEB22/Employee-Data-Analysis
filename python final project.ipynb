3. Predominant Age Group
🔹 Employees in the [most frequent age group] (e.g., 26-35 years) form the largest segment of the workforce.
🔹 The presence of younger employees might indicate recent hiring trends, while fewer older employees may suggest retirements or transitions to leadership roles.

📌 Business Impact: This insight can help HR teams with succession planning, training programs, and employee retention strategies.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame (Replace this with your actual data)
data = {
    "age": [22, 25, 28, 35, 40, 45, 50, 55, 60, 30, 34, 38, 42, 48, 52, 58, 62]
}
df = pd.DataFrame(data)

# Define age bins and labels
age_bins = [18, 25, 35, 45, 55, 65]  # Age bin ranges
age_labels = ["18-25", "26-35", "36-45", "46-55", "56-65"]  # Labels

# Categorize ages into bins
df["age_group"] = pd.cut(df["age"], bins=age_bins, labels=age_labels, right=False)

# Count occurrences of each age group
age_group_counts = df["age_group"].value_counts().sort_index()

# Find the predominant age group
predominant_age_group = age_group_counts.idxmax()
print(f"Predominant Age Group: {predominant_age_group}")

# Visualization
plt.figure(figsize=(8, 5))
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette="magma")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.title("Predominant Age Group")
plt.xticks(rotation=45)
plt.show()
```

4. Salary Expenditure by Team & Position
🔹 The [team with highest salary] has the highest total salary expenditure, likely due to a larger team size or highly paid specialized roles.
🔹 The [position with highest salary] consumes the most salary resources, which may indicate executive-level positions or technical expertise being rewarded.

📌 Business Impact: The finance department can use this data to optimize salary budgets and ensure fair compensation structures.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame (Replace with actual data)
data = {
    "team": ["A", "B", "A", "C", "B", "A", "C", "B", "C", "A"],
    "position": ["Manager", "Analyst", "Developer", "Manager", "Developer", 
                 "Analyst", "Manager", "Developer", "Analyst", "Manager"],
    "salary": [80000, 60000, 70000, 90000, 75000, 62000, 85000, 78000, 70000, 88000]
}
df = pd.DataFrame(data)

# Ensure column names are clean
df.columns = df.columns.str.strip()

# Check if required columns exist
required_columns = {"team", "position", "salary"}
if required_columns.issubset(df.columns):

    # Group by Team and Position
    team_salary = df.groupby("team")["salary"].sum().sort_values(ascending=False)
    position_salary = df.groupby("position")["salary"].sum().sort_values(ascending=False)

    # Team Salary Visualization
    plt.figure(figsize=(8, 5))
    sns.barplot(x=team_salary.index, y=team_salary.values, palette="Blues_r")
    plt.xticks(rotation=45)
    plt.xlabel("Team")
    plt.ylabel("Total Salary")
    plt.title("Total Salary Expenditure per Team")
    plt.show()

    # Position Salary Visualization
    plt.figure(figsize=(8, 5))
    sns.barplot(x=position_salary.index, y=position_salary.values, palette="Reds_r")
    plt.xticks(rotation=45)
    plt.xlabel("Position")
    plt.ylabel("Total Salary")
    plt.title("Total Salary Expenditure per Position")
    plt.show()

else:
    missing = required_columns - set(df.columns)
    print(f"Missing columns: {missing}. Please check your dataset.")
```
